```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and contains 'gender' and 'math_score' columns
# Calculate the average math_score for each gender
avg_math_score = df.groupby('gender')['math_score'].mean().reset_index()

# Plotting the average math score for male and female
plt.figure(figsize=(6, 4))
sns.barplot(x='gender', y='math_score', data=avg_math_score,hue='gender', palette='viridis', width=0.2)
plt.title('Average Math Score by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Average Math Score', fontsize=14)

plt.show()
```

```python
# Select only numeric columns (like 'average', 'math_score')
numeric_columns = ['average', 'math_score']

# Group the data by gender and calculate mean only for numeric columns
gender_group = df.groupby('gender')[numeric_columns].mean()

# Create figure and axes
# plt.figure(figsize=(10, 8))

# Define categories for X-axis
X = ['Total Average', 'Math Average']

# Extract the average and math scores for male and female
female_scores = [gender_group.loc['female', 'average'], gender_group.loc['female', 'math_score']]
male_scores = [gender_group.loc['male', 'average'], gender_group.loc['male', 'math_score']]

# Set up X-axis positions
X_axis = np.arange(len(X))

# Plot bars for male and female scores
plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male', color='blue')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female', color='pink')

# Add labels and title
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total Average vs Math Average Marks of Male and Female Students", fontweight='bold')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Select only the numeric columns
numeric_columns = df.select_dtypes(include='number')

# Group the data by 'race_ethnicity' and calculate the mean of numeric columns
Group_data2 = df.groupby('race_ethnicity')[numeric_columns.columns].mean()

# Create subplots with 1 row and 3 columns
f, ax = plt.subplots(1, 3, figsize=(20, 8))

# Plot for Math Score
sns.barplot(x=Group_data2.index, y=Group_data2['math_score'],hue=Group_data2.index, palette='mako', ax=ax[0])
ax[0].set_title('Math score', color='#005ce6', size=20)

# Add bar labels
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=15)

# Plot for Reading Score
sns.barplot(x=Group_data2.index, y=Group_data2['reading_score'],hue=Group_data2.index, palette='flare', ax=ax[1])
ax[1].set_title('Reading score', color='#005ce6', size=20)

# Add bar labels
for container in ax[1].containers:
    ax[1].bar_label(container, color='black', size=15)

# Plot for Writing Score
sns.barplot(x=Group_data2.index, y=Group_data2['writing_score'],hue=Group_data2.index, palette='coolwarm', ax=ax[2])
ax[2].set_title('Writing score', color='#005ce6', size=20)

# Add bar labels
for container in ax[2].containers:
    ax[2].bar_label(container, color='black', size=15)

# Display the plot
plt.show()
```
